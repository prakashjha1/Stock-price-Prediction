```python
from sklearn.model_selection import train_test_split # it will help to split dataset
from sklearn.linear_model import LinearRegression 
from sklearn import metrics
import math
import pandas as pd # it will help to load or manipulate dataset
import matplotlib.pyplot as plt # it will help to plot graph
%matplotlib inline
```

```python
#importing or loading dataset using pandas library
dataset = pd.read_csv("/home/samar/Downloads/NSE-TATAGLOBAL11.csv")
```

```python
dataset.describe()
```

```python
# Here x will represent as inputs and y will be our output (which we want to predict)
x = dataset[['High','Low','Open','Total Trade Quantity']].values
y = dataset[['Close']].values
```

```python
# Here we will split our dataset into test dataset and train dataset
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Here we will create object of LinearRegression
regressor = LinearRegression()
```

```python
# Let's train the model by giving x_train and y_train 
regressor.fit(x_train,y_train)
```

```python
# Let's find the value of coefficient and intercept
# y = mx + c , here m is coefficent or slope and c is intercept
print("Values of coefficient =>",regressor.coef_)
print("Value of Intercept =>",regressor.intercept_)
```

```python
# Here we will predict the value
predicted = regressor.predict(x_test)
```

```python
# we cannot easily understand the predicted value by printing it
print(predicted)
```

```python
# Here we can create a dataframe which can show actual values and predicted values
dataframe = pd.DataFrame({'Actual value':y_test.flatten(),'Predicted value':predicted.flatten()})
```

```python
# Here we will print first 20 rows of a our dataframe
dataframe.head(20)
```

```python
# Here we see accuracy of our model by calculating the errors(less error, more accurate )
mae = metrics.mean_absolute_error(y_test,predicted)
mse = metrics.mean_squared_error(y_test,predicted)
rmse = math.sqrt(mse)
print("Mean Absolute Error : ",mae)
print("Mean Squared Error : ",mse)
print("Root Mean Squared Error : ",rmse)
```

```python
# Now we will plot graph for actual value and predicted value of first 20 rows of dataset
graph = dataframe.head(20)
graph.plot(kind="bar") # here we are plotting bar graph to visualize dataframe
```
